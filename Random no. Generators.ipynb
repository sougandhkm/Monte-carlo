```python
import random
from numpy import random as nprandom
import matplotlib.pyplot as plt
from timeit import default_timer as timer
```

```python
#random.seed(1)
#nprandom.seed(1)
```

```python
randuni=[]
npuni=[]
step=[]
for i in range (0,1000):
    randuni.append(random.uniform(0,1))
    npuni.append(nprandom.uniform(0,1))
    step.append(i)
print("The random number generated points")
plt.xlabel('steps')
plt.ylabel('numbers')
plt.plot(step,randuni,'o')
plt.show()
print("the numpy random no. genrated points" )
plt.xlabel('steps')
plt.ylabel('numbers')
plt.plot(step,npuni,'o')
plt.show()
```

```python
start=timer()
for i in range(0,1000):
    random.uniform(0,10)
end=timer()
print("The time to generate 1000 uniform points by python random generator is =" ,(end-start))
start=timer()
for i in range(0,1000):
    nprandom.uniform(0,10)
end=timer()
print("The time to generate 1000 uniform points by numpy random generator is =" ,(end-start))
start=timer()
for i in range(0,1000):
    random.randint(0,10)
end=timer()
print("The time to generate 1000 integer points by python random generator is =" ,(end-start))
start=timer()
for i in range(0,1000):
    random.randint(0,10)
end=timer()
print("The time to generate 1000 integer points by numpy random generator is =" ,(end-start))
start=timer()
```
