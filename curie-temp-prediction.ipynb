```python
#import essential libraries
import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
#importing model
model=keras.models.load_model('curietemp4.h5')
```

```python
#Creating temperature data to predict Binder's cumulant for different lattices
Temp_x=np.linspace(1.9,2.6,3000)
N_x=np.tile(28,3000)
N_x1=np.tile(32,3000)
N_x2=np.tile(40,3000)
```

```python
#Reshaping the data to give as input
Test_X=np.array([[Temp_x],[N_x]])
Test_X1=np.array([[Temp_x],[N_x1]])
Test_X2=np.array([[Temp_x],[N_x2]])
```

```python
Test_X=Test_X.reshape(2,3000)
Test_X=np.transpose(Test_X)
Test_X1=Test_X1.reshape(2,3000)
Test_X1=np.transpose(Test_X1)
Test_X2=Test_X2.reshape(2,3000)
Test_X2=np.transpose(Test_X2)
```

```python
#Prediction using loaded model
Test_Y=model.predict(Test_X)
Test_Y1=model.predict(Test_X1)
Test_Y2=model.predict(Test_X2)
```

```python
#plotting the results
plt.plot(np.transpose(Temp_x),Test_Y)
plt.plot(np.transpose(Temp_x),Test_Y1)
plt.plot(np.transpose(Temp_x),Test_Y2)
plt.xlabel('Temperature')
plt.ylabel('Binder cumulant')
legend=['L=28','L=32','L=40']
plt.legend(legend)
plt.show()
```
